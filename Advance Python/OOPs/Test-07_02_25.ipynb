```python
#Synchronous Program Example
import time

class SynchronousExample:
  def task1(self):
    print("Starting Task 1....")
    time.sleep(4)
    print("Task 1 Completed")

  def task2(self):
    print("Starting Task 2....")
    time.sleep(4)
    print("Task 2 Completed")

syncObj=SynchronousExample()

syncObj.task1()
syncObj.task2()
```

```python
#Asynchronous Program Example
import asyncio

class AsynchronousExample:
  async def task1(self):
    print("Starting Task 1....")
    await asyncio.sleep(3)
    print("Task 1 Completed")

  async def task2(self):
    print("Starting Task 2....")
    await asyncio.sleep(3)
    print("Task 2 Completed")

  async def main(self):
    await asyncio.gather(self.task1(),self.task2())

asyncObj=AsynchronousExample()

await asyncObj.main()
```
