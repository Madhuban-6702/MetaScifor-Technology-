```python
#Single Inheritance
class Parent:
  def P1(self):
    print("Parent!!")

class Child(Parent):
  def C1(self):
    print("Child")

child=Child()
child.P1()
child.C1()
```

```python
#Multiple Inheritance
class Father:
  def Dad(self):
    print("The father!!")

class Mother:
  def Mom(self):
    print("The mother!!")

class Child(Father, Mother):
  def kid(self):
    print("The Child")

child=Child()
child.Dad()
child.Mom()
child.kid()
```

```python
#Multilevel Inheritance
class GrandFather:
  def GP(self):
    print("Grandfather!!!")

class Father(GrandFather):
  def Dad(self):
    print("Father!!!")

class Son(Father):
  def Kid(self):
    print("Child!!!")

child=Son()
child.GP()
child.Dad()
child.Kid()
```

```python
#Hierarchical Inheritance
class Father:
  def Dad(self):
    print("The father!!")

class Son(Father):
  def _son(self):
    print("Son!!!")

class Daughter(Father):
  def _daughter(self):
    print("Daughter!!!")

son=Son()
son.Dad()
son._son()

print()

daughter=Daughter()
daughter.Dad()
daughter._daughter()
```

```python
#Polymorphism
class Car:
  def model(self):
    print("BMW")

class Bike:
  def model(self):
    print("Z900")

def vehicle_model(vehicle):
  vehicle.model()

vehicle_model(Car())
vehicle_model(Bike())
```

```python
#Encapsulation
class Person:
  def __init__(self,name):
    self.__name = name

  def getName(self):
    return self.__name

  def setName(self,newName):
    self.__name = newName

p=Person("Madhuban")
print(p.getName())
print()
p.setName("Maddy")
print(p.getName())
```

```python
#Abtraction
from abc import ABC, abstractmethod
class Vehicle(ABC):
  @abstractmethod
  def startEngine(self):
    pass

class Car(Vehicle):
  def startEngine(self):
    print("Start with key ignition.")

class Bike(Vehicle):
  def startEngine(self):
    print("Starts with selfstart button on Kick.")

car=Car()
car.startEngine()
print()
bike=Bike()
bike.startEngine()
```
