Q2. Himanshi has a text file and she wants to check whether that file contains any numerical value. So to help Himanshi write a python program that can check whether the text file includes any numerical value or not. As text files contain so many lines, so print the line number that has numerical values

```python
f=open("/content/sample_data/Q2.txt","r")
lines=f.readlines()
print("Lines containing numerical values: ")
for i, line in enumerate(lines, start=1):
  if any(char.isdigit() for char in line):
    print(f"Line {i}: {line.strip()}")
f.close()
```

Q3. Shubham had created a BMI calculator to check his health condition daily. But he is not able to record the values. He wants to see how his BMI is changing. So he decided to connect his BMI program with a text file using the concept of file handling. Now, when he uses his BMI program, his height, weight, BMI, and date get added to a text file "bmi.txt". Create a program that can do the same thing for you.

```python
def calculate_bmi():
  height =float(input("Enter Your height in meters: "))
  weight=float(input("Enter your weight in kilograms: "))

  bmi=weight/(height**2)

  f=open("/content/sample_data/bmi.txt","a")
  f.write(f"Height: {height}m, Weight: {weight}Kg, BMI: {bmi:.2f}\n")
  print(f"Your BMI is: {bmi:.2f}")
  print("Your data has been saved to bmi.txt file!!")
  f.close()

calculate_bmi()
```

Q4. Arvind has a huge text file and he wants to put a serial number at the beginning of each line. Doing this task manually can take a lot of time. He is not aware of python file handling. Write a python program that can put a serial number in front of each line in any specified text file.


```python
f=open("/content/sample_data/aboutPython.txt","r")
lines=f.readlines()

file=open("/content/sample_data/aboutPython.txt","w")
for i, line in enumerate(lines,start=1):
  file.write(f"{i}.{line}")

print("Serial Number Added Successfully!!!")
```

Q7. Nidhi has a bad habit of repeating the same word while typing. She wants a program that can remove duplicate phrases from paragraphs written by her. To help Nidhi write a python program that can connect with a text file and remove repeating words.


```python
f=open("/content/sample_data/Q7.txt","r")
text=f.read()
print(text)
print()

words=text.split()
cleaned_words=[]

for i in range(len(words)):
  if i == 0 or words[i] != words[i-1]:
    cleaned_words.append(words[i])

cleaned_text=" ".join(cleaned_words)

file=open("/content/sample_data/Q7-output.txt","w")
file.write(cleaned_text)
print("Duplicate Phrases removed successfully.\nYou can check the outputfile")
print()
print("Cleaned Text:")
print(cleaned_text)
```
