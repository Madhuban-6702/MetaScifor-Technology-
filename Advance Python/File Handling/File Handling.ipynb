```python
#Read Mode(r)
f=open("/content/sample_data/sample.txt","r")
print("Reading file content: ")
print(f.read())
f.close()

print()

f=open("/content/sample_data/sample.txt","r")
print("Reading 1st line from file: ")
print(f.readline())
f.close()
```

```python
#Writing Mode(w)
f=open("/content/sample_data/sample.txt","w")
f.write("Let's demonstrate the file handling using write mode in python.\n")
f.close()

f=open("/content/sample_data/sample.txt","r")
print("After uing Write mode: ")
print(f.read())
f.close()
```

```python
#Append Mode(a)
f=open("/content/sample_data/sample.txt","a")
f.write("Appending the lines. \nHello My name is Madhuban.")
f.close()

f=open("/content/sample_data/sample.txt","r")
print("After using Append mode: ")
print(f.read())
f.close()
```

```python
#Read and Write mode(+r)
f=open("/content/sample_data/sample.txt","r+")
print("Befor Changing: ",f.read())
f.seek(0)
f.write("Overwriting the content.\n")
f.close()

print()

f=open("/content/sample_data/sample.txt","r")
print("After Changing: ")
print(f.read())
f.close()
```

```python
#Write and Read Mode(w+)
f=open("/content/sample_data/sample.txt","w+")
f.write("Completely new content written.\n")
f.seek(0)
print("After using write and read mode: ")
print(f.read())
f.close()
```

```python
#Append and Read mode(a+)
f=open("/content/sample_data/sample.txt","a+")
f.write("Adding content in append and read mode.\n")
f.seek(0)
print("after using append and read mode: ")
print(f.read())
f.close()
```

```python
#Count the number of Special Characters , lowercase , uppercase and number in file
special_char = num = lowercase = uppercase = 0
with open("/content/sample_data/TestFile.txt","r") as file:
  content = file.read()
  print("File Content: ", content)
  print()

  for char in content:
    if char.islower():
      lowercase += 1
    elif char.isupper():
      uppercase += 1
    elif char.isdigit():
      num += 1
    elif char.isspace():
      special_char += 1

print(f"Lowercase Letters: {lowercase}")
print(f"Uppercase Letters: {uppercase}")
print(f"Numbers: {num}")
print(f"Special Character: {special_char}")
f.close()

```
