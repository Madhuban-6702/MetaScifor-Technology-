```python
#Get Method
import json
import requests

url="https://jsonplaceholder.typicode.com/posts/1/comments"

res = requests.get(url)

if res.status_code == 200:
  print("Get Response: \n",res.json())
else:
  print("Failed to fetch data, Status Code: ",res.status_code)
```

```python
#Post Method
import json
import requests

url = "https://jsonplaceholder.typicode.com/posts"

post_data={
    "postId" : 1,
    "id" : 1,
    "name" : "New Comment",
    "email" : "newContent@gmail.com",
    "body" : "This is a newly created comment"
}
res=requests.post(url,json=post_data)

if res.status_code == 201:
  print("Post Response: \n",res.json())
else:
  print("Failed to create a post, Status Code: ", res.status_code)
```

```python
#Put Method
import json
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"

update_data={
    "postId" : 1,
    "id" : 1,
    "name" : "Update Comment",
    "email" : "updateContent@gmail.com",
    "body" : "This comment has been updated."
}

res=requests.put(url,json=update_data)

if res.status_code == 200:
  print("Put Response: \n",res.json())
else:
  print("Failed to update the comment, Status Code: ", res.status_code)
```

```python
#Delete Method
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"

res=requests.delete(url)

if res.status_code == 200:
  print("Delete Response: \nComment deleted successfully.")
else:
  print("Failed to delete the comment, Status Code: ".res.status_code)
```

```python
#Json module data using json load method
import json
employees="""
{
  "emp" :[
    {
      "name" : "Madhuban",
      "emp_id" : "E101",
      "email" : "madhuban22@gmail.com"
    },
    {
      "name" : "Aditya",
      "emp_id" : "E102",
      "email" : "Adi102@gmail.com"
    },
    {
      "name" : "Saurabh",
      "emp_id" : "E103",
      "email" : "Saurabh67@gmail.com"
    }
  ]
}
"""

data=json.loads(employees)
print(data['emp'])
print(type(data['emp']))

for employee in data['emp']:
  print(employee)
```

```python
#Json module data using json dump method
import json
employees="""
{
  "emp" :[
    {
      "name" : "Madhuban",
      "emp_id" : "E101",
      "email" : "madhuban22@gmail.com",
      "Dept" : "IT"
    },
    {
      "name" : "Aditya",
      "emp_id" : "E102",
      "email" : "Adi102@gmail.com",
      "Dept" : "Tech Support"
    },
    {
      "name" : "Saurabh",
      "emp_id" : "E103",
      "email" : "Saurabh67@gmail.com",
      "Dept" : "CRM"
    }
  ]
}
"""

data=json.loads(employees)
print(data['emp'])
print(type(data['emp']))

for employee in data['emp']:
  del employee['Dept']

new_employee = json.dumps(data,indent=3,sort_keys=True)
print(new_employee)
```
