Q1. Create a student management using dictionary, list and use it inside a loops

```python
students={}

while True:
  print("\n--Student Management System--")
  print("1. Add Students")
  print("2. View all students")
  print("3. Update a student")
  print("4. Delete a student")
  print("5. Exit")

  choice=input("Enter Your Choice: ")

  if choice == "1":
    id=int(input("Enter Student ID: "))
    name=input("Enter Student Name: ")
    students[id]=name
    print("Student Added")

  elif choice == "2":
    if students:
      for id, name in students.items():
        print(f"ID: {id}, Name:{name}")
    else:
      print("No Student Found.")

  elif choice == "3":
    id=int(input("Enter Student ID to update: "))
    if id in students:
      students[id]=input("New Name: ")
      print("Student Updated!!")
    else:
      print("Student Not found.")

  elif choice == "4":
    id=int(input("enter Student ID yo delete: "))
    if id in students:
      del students[id]
      print("Student Deleted!!")
    else:
      print("Student Not found.")

  elif choice == "5":
    print("Exit")
    break

  else:
    print("Invalid Input")
```

Q2. Print and find prime numbers 1 to 100

```python
print("Prime number between 1 to 100")
for num in range(2, 101):
  isPrime=True
  for i in range(2,int(num ** 0.5)+1):
    if num % i == 0:
      isPrime=False
      break
  if isPrime:
    print(num, end=" ")
```

Q3. Create a fibonnaci series

```python
num=int(input("Enter the number of terms for Fibonnaci Series: "))
a,b = 0, 1
print("Fibonnaci Series")
for _ in range(num):
  print(a, end=" ")
  a,b=b,a+b
```

Q4. Create the Factorial program

```python
#Using For Loop
num=int(input("Enter a number to find the factorial for it: "))
fact=1
for i in range(1,num+1):
  fact *= i
print(f"The factorial of {num} is {fact}")
```

```python
#Using While loop
num=int(input("Enter a number to find the factorial for it: "))
fact=1
i=1

while i<=num:
  fact *= i
  i+=1

print(f"The factorial of {num} is {fact}")
```

Q5. Print even and odd numbers from 1 to 100

```python
print("Even number between 1 to 100")
for i in range(1,101):
  if i%2 == 0:
    print(i,end=" ")

print()

print("\nOdd Number between 1 to 100")
for i in range(1,101):
  if i%2 != 0:
    print(i,end=" ")
```

Q6. Create a quiz program with total number of correct and the wrong answers

```python
ques={
    "What is 2+2?": "4",
    "What is the color of the sky?": "Blue",
    "What is the largest planet in our solar system?": "Jupiter"
}

correct=0
wrong=0

print("\nWelcome to the Quiz Game!!")
print("Answer the questions.\n")

for ques,ans in ques.items():
  print(ques)
  answer=input("Your Ans: ").strip()
  if answer.lower() ==  ans.lower():
    print("Correct!!!\n")
    correct += 1
  else:
    print("\nWrong!!!\n")
    wrong += 1

print(f"Quiz Over!! You got {correct} correct and {wrong} wrong answers.")
```
