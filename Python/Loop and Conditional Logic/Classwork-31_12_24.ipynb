Q1. Create a sum of 100 numbers program using for loop and the while loop separately

```python
#For Loop
sum_of_100=0
for i in range(1,101):
  sum_of_100 += i
print("Sum of 100 numbers: ", sum_of_100)
```

```python
#While Loop
sum_of_100=0
i=1
while i<=100:
  sum_of_100 += i
  i+=1
print("Sum of 100 numbers: ", sum_of_100)
```

Q2. Create a quiz program using while loop

```python
questions=[("What is the color of the sky?","blue"),
           ("What is 4+4?","8"),
           ("Which is the largest planet in solar system","Jupiter")]

i=0
score=0

while i<len(questions):
  q, ans=questions[i]
  if input(q + " ").strip().lower() == ans:
    print("Correct!!")
    score += 1
  else:
    print("Wrong!!")
  i+=1
print(f"You answered {score} out of {len(questions)} questions correctly!!")
```

Q4. Create ATM Program

```python
balance=1000
print("Welcome to the ATM!!")
while True:
  print("\n1. Check Balance\n2. Deposit Money\n3. Withdraw Money\n4. Exit")
  choice=input("Enter your choice: ")

  if choice == "1":
    print(f"Your current balance is ${balance}")
  elif choice == "2":
    deposit=int(input("Enter the amount to deposit: "))
    balance += deposit
    print(f"You have successfully deposited ${deposit}. Your new balance is ${balance}")
  elif choice == "3":
    withdraw = int(input("Enter the amount to withdraw: "))
    if withdraw <= balance:
      balance -= withdraw
      print(f"You hae successfully withrawn ${withdraw}. Your remaining balance is ${balance}")
    else:
      print("Insufficient Balance!!")
  elif choice == "4":
    print("Thnak you!!")
    break
  else:
    print("Invalid Choice!!\nPlease try again!!")
```
