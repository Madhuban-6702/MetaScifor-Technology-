```python
#Create a Simple Calculator using functions
def cal():
  print("Calculator")
  while True:
    print("\n1. Add\n2. Subtract\n3. Multiplication\n4. Divide\n5. Exit")
    choice=input("Enter your choice: ")
    a=int(input("Enter 1st number: "))
    b=int(input("Enter 2nd number: "))
    if choice == "1":
      print("Addition is: ",a+b)
    elif choice == "2":
      print("Subtraction is: ",a-b)
    elif choice == "3":
      print("Multiplication is: ",a*b)
    elif choice == "4":
      print("Division is: ",a/b)
    elif choice== "5":
      print("Exit")
      break
    else:
      print("Invalid input")

cal()
```

```python
#Quiz Program
def question(ques,correct_ans):
  ans=input(ques + " ").strip().lower()
  if ans==correct_ans:
    print("Correct!!")
    return 1
  else:
    print("Wrong!!")
    return 0

def quiz():
  questions=[
      ("What is 4+4?","8"),
      ("What is the capital of france?","paris"),
      ("What is the color of the sky?","blue")
  ]

  score=0
  for ques, correct_ans in questions:
    score += question(ques,correct_ans)

  print(f"you answered {score} out of {len(questions)} questions correctly!!")

quiz()
```
