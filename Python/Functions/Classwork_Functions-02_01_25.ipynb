Q1. Pastel Text Box
Chirag was given a list containing n numbers. His teacher asked him to find the product of all the elements present in the list. He did it perfectly when the list was small. As the size of the next lists were becoming larger, he found it difficult to calculate the product. He thought of creating a python program that gets a list as argument and calculates the product and displays the result. But he is not good at function concepts. Help him to create a `function` to do the above.

```python
def calculate_product(num):
  product = 1
  for i in num:
    product *= i
  return product

product_list=[1,2,3,4,5,6,7]
result=calculate_product(product_list)
print("The product of the list is: ",result)
```

Q2. Sahil came to know about Pascal’s triangle. He tries the small patterns of pascal’s triangle. But he was not sure how the triangle looks when the number of rows were increased. Help him to create a function which gets the number of rows as input from the user and prints the pascal’s triangle upto that row.

```python
def pascal_triangle(rows):
  triangle=[]
  for i in range(rows):
    row=[1]
    if triangle:
      last_row=triangle[-1]
      row.extend([last_row[j]+last_row[j+1] for j in range(len(last_row)-1)])
      row.append(1)
    triangle.append(row)
  return triangle

num_row=int(input("Enter the number of rows for pascal's triangle: "))
pascals_tri=pascal_triangle(num_row)

for row in pascals_tri:
  print(" ".join(map(str,row)).center(num_row*2))
```

Q3. Retro Frame Text Box
Nidhi is writing a program to count the number of digits in a number given by user. But she got stuck in the middle. Help her to create a function that counts the number of digits in a given number.

```python
def count_digit(num):
  return len(str(abs(num)))

number=int(input("Enter a number: "))
count_num=count_digit(number)
print(f"The number of digit in {number} is: {count_num}")
```

Q4. Mahesh learnt about LCM of numbers and is
eager to find LCM for more numbers. But he
finds it tough to calculate the LCM for larger
numbers. Help him to create a function that
gets two numbers as input, finds the LCM of
those two numbers and displays the result as
output.

```python
def find_lcm(num1,num2):
  greater=max(num1,num2)

  while True:
    if greater % num1 == 0 and greater % num2 == 0:
      return greater
    greater += 1

number1=int(input("Enter the first number: "))
number2=int(input("Enter the second number: "))

lcm=find_lcm(number1,number2)
print(f"The LCM of {number1} and {number2} is: {lcm}")
```
